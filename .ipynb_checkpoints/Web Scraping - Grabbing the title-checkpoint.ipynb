```python
import requests
```

```python
result = requests.get("http://www.example.com")
```

```python
type(result)
```

```python
result.text
```

```python
import bs4
```

```python
#To convert it into a soup
#We will use beautiful suup
```

```python
soup = bs4.BeautifulSoup(result.text,"lxml")
```

```python
soup
```

```python
#this is for selecting the title
soup.select("title")
```

```python
#similarly for selecting the body
soup.select("body")
```

```python
#or for paragraphs
soup.select("p")
```

# For actual text

```python
#Title text
soup.select("title")[0].getText()
```

```python
#Another method to grap text
site_para = soup.select("p")
```

```python
site_para[0].getText()
```

# similarly for another website

```python
site2 = requests.get("https://en.wikipedia.org/wiki/Narendra_Modi")
```

```python
type(site2)
```

```python
site2.text
```

```python
grab_elem = bs4.BeautifulSoup(site2.text,"lxml") 
```

```python
#for page source code
grab_elem
```

```python
grab_elem.select("title")
```

```python
#selecting only title text
only_title = grab_elem.select("title")
```

```python
only_title[0].getText()
```

```python
#selecting only paragraph text
only_para = grab_elem.select("body")
```

```python
only_para[0].getText()
```
